# Data Preprocessing - Delhi Traffic Dataset

```python
import pandas as pd

# Load datasets
features_df = pd.read_csv("../data/raw/delhi_traffic_features.csv")
target_df = pd.read_csv("../data/raw/delhi_traffic_target.csv")

# Preview
features_df.head()
```

```python
# Check duplicate rows
features_df.duplicated().sum()
```

```python
# Merge features and target data
traffic_df = pd.merge(features_df, target_df, on="Trip_ID")

# Preview merged dataset
traffic_df.head()
```

```python
# Save cleaned/merged dataset
traffic_df.to_csv(
    "../data/cleaned/delhi_traffic_cleaned.csv",
    index=False
)

print("Cleaned dataset saved successfully!")
```
